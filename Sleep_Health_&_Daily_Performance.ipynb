```python
#Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in data.select_dtypes(include='object').columns:
    data[col] = le.fit_transform(data[col])
```

Target - felt_rested


```python
#Checking if data is balanced or not for first target variable (felt_rested)
data['felt_rested'].value_counts()
```

```python
#Train-Test-Split
from sklearn.model_selection import train_test_split

x = data.drop('felt_rested', axis=1)
y = data['felt_rested']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#Model 1: Logistic Regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 100, class_weight='balanced')
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score

# Get probability for class 1
y_prob = model.predict_proba(x_test)[:, 1]

# Calculate AUC
auc = roc_auc_score(y_test, y_prob)

print("AUC:", auc)
```

```python
#Model 2: Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred_rf = rf.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import roc_auc_score

# Get probability for class 1
y_prob = rf.predict_proba(x_test)[:, 1]

# Calculate AUC
auc = roc_auc_score(y_test, y_prob)

print("AUC:", auc)
```

```python
#Model 3: XGBoost
from xgboost import XGBClassifier

xgb = XGBClassifier(
    eval_metric = 'logloss'
)

xgb.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

y_pred_xgb = xgb.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print(classification_report(y_test, y_pred_xgb))
```

```python
from sklearn.metrics import roc_auc_score

# Get probability for class 1
y_prob = xgb.predict_proba(x_test)[:, 1]

# Calculate AUC
auc = roc_auc_score(y_test, y_prob)

print("AUC:", auc)
```

Conclusion - Logistic Regression/Random Forest both performed best with AUC: 0.82

Target- cognitive_performance_score

```python
i = data.drop(['cognitive_performance_score'], axis=1)
j = data['cognitive_performance_score']
```

```python
#Train-Test-Split
from sklearn.model_selection import train_test_split

i_train, i_test, j_train, j_test = train_test_split(i, j, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

i_train = scaler.fit_transform(i_train)
i_test = scaler.transform(i_test)
```

```python
#Model 1: Linear Regression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(i_train, j_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

j_pred = model.predict(i_test)

rmse = np.sqrt(mean_squared_error(j_test, j_pred))
mae = mean_absolute_error(j_test, j_pred)

print("RMSE:", rmse)
print("MAE:", mae)
```

```python
#Model 2: Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(i_train, j_train)
```

```python
j_pred_rf = rf.predict(i_test)

rmse_rf = np.sqrt(mean_squared_error(j_test, j_pred_rf))
print("RF RMSE:", rmse_rf)
```

```python
#Model 3: XGBoost
from xgboost import XGBRegressor

xgb = XGBRegressor(
    n_estimators=100,
    random_state=42
)

xgb.fit(i_train, j_train)
```

```python
j_pred_xgb = xgb.predict(i_test)

rmse_xgb = np.sqrt(mean_squared_error(j_test, j_pred_xgb))
print("XGB RMSE:", rmse_xgb)
```

Conclusion- XGBoost performed best with RMSE: 5.99
